```python
# dependencies 
import requests
import gmaps

#import API Key
from config import g_key
```

```python
# set the parameters to search for a hotel in Paris

params = {
    "radius":5000,
    "types":'lodging',
    "key" :g_key,
    "location" :"48.8566, 2.3522"
}

# Use the base URL to search for the hotels in Paris
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

#Make requests and get he JSON data from the scratch
hotels =requests.get(base_url,params=params).json()
hotels
```
